```python
class Stack:
    def __init__(self):
        self.stack = []
    
    def push(self, item):
        self.stack.append(item)  # O(1)
        print(f"Pushed {item} to stack")

    def push_multiple(self, items):
        for item in items:
            self.push(item)

    def pop(self):
        if self.is_empty():
            return "Stack is empty"
        item = self.stack.pop()  # O(1)
        return item
    
    def peek(self):
        if self.is_empty():
            return "Stack is empty"
        return self.stack[-1]  # Return last element
    
    def is_empty(self):
        return len(self.stack) == 0
    
    def size(self):
        return len(self.stack)
    
    def display(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            print("Top ->", self.stack[::-1])  # Display top to bottom

# Usage
s = Stack()
s.push(1)
s.push(2)
s.push(3)
s.display()      # Top -> [3, 2, 1]
print(s.pop())   # 3
print(s.peek())  # 2
s.push_multiple([2, 4, 5, 7])  # â† Fixed here
s.display()      # Top -> [2, 1]
```
